```python
# Dependencies and Setup
from pandas import DataFrame
import pandas as pd
import numbers as num

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_data["SN"].unique())

player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item Name"].unique())
average_price = (purchase_data["Price"].sum()) / (purchase_data["Purchase ID"].count())

print(unique_items)
print(average_price)

summary_df = pd.DataFrame({"# of Unique Items":[unique_items], "Average Price":[average_price]})
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_demo = purchase_data.loc[:, ["Gender", "SN", "Age"]]
gender_demo = gender_demo.drop_duplicates()

gender_count = gender_demo["Gender"].value_counts()
gender_count_percent = gender_count / player_count * 100
gender_df = pd.DataFrame({"Total": gender_count, "Percent": gender_count_percent})
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_total = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase value")
average_price = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Price")
purchase_count = purchase_data.groupby(["Gender"]).count()["Price"].rename("Total Purchase Count")

purchasing_analysis_df = pd.DataFrame({"Purchase Total": purchase_total, "Average Price": average_price, "Purchase Count": purchase_count})
purchasing_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#age_bin = purchase_data.groupby("Age")
#pd.DataFrame(age_bin)
bins = [0,16,29,100]
bin_names = ["Minor", "Early Adult", "Mid-Upper Adult"]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=bin_names)
age_group_count = purchase_data.groupby("Age Group")


average_age = purchase_data.groupby(["Age Group"]).mean()["Age"].rename("Average Age")
category_count = purchase_data.groupby(["Age Group"]).count()["Age"].rename("Total in Category")
age_count_percent = category_count / purchase_data["Age Group"].count() * 100

purchasing_analysis_df = pd.DataFrame({"Average Age": average_age, "% of Total": age_count_percent, "Total in Category": category_count})
purchasing_analysis_df

#category_count
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=bin_names)
age_group_count = purchase_data.groupby("Age Group")

purchase_total = purchase_data.groupby(["Age Group"]).sum()["Price"].rename("Total Purchase value")
average_price = purchase_data.groupby(["Age Group"]).mean()["Price"].rename("Average Price")
purchase_count = purchase_data.groupby(["Age Group"]).count()["Price"].rename("Total Purchase Count")

purchasing_analysis_df = pd.DataFrame({"Purchase Total": purchase_total, "Average Price": average_price, "Purchase Count": purchase_count})
purchasing_analysis_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_price = purchase_data.groupby(["Item Name"]).mean()["Price"].rename("Item Price")
purchase_count = purchase_data.groupby(["Item Name"]).count()["Price"].rename("Total Purchase Value")
#purchase_total = purchase_data.groupby(["Item Name"]).sum()["Price"].rename("Total Purchase value")
purchase_total = item_price * purchase_count

purchasing_analysis_df = pd.DataFrame({"Average Item Price": item_price, "Purchase Count": purchase_count, "Purchase Totals": purchase_total})
purchasing_analysis_df
```
